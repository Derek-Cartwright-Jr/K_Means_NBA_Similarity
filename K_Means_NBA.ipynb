```python
#First we need to create the rowset and column headers for our dataframe.
players = []
season_stats = {'name':None, 'PTS':None, 'REB':None, 'AST':None, 'STL':None, 'BLK':None}

def get_stats(name, player_id):
    player_url = 'http://stats.nba.com/stats/playercareerstats?'+\
                 'LeagueID=00&PerMode=PerGame&PlayerID='+player_id

    response = requests.get(player_url)
    data = json.loads(response.text)
    stats = data['resultSets'][0]['rowSet'][-1:]
    headers = data['resultSets'][0]['headers']

    df = pd.DataFrame(stats, columns=headers)

    season_stats['name'] = name
    season_stats['PTS'] = df['PTS'].item()
    season_stats['REB'] = df['REB'].item()
    season_stats['AST'] = df['AST'].item()
    season_stats['BLK'] = df['BLK'].item()
    season_stats['STL'] = df['STL'].item()

    players.append(season_stats.copy())

for team in teams:
    for player in teams[team]:
        get_stats(player,teams[team][player])

column_heads = ['name', 'PTS', 'AST', 'REB', 'STL', 'BLK']
Player_df = pd.DataFrame(players, columns=column_heads)

#Dataframe of each players points, assists, rebounds, blocks, and steals per game.
print(Player_df)
```

```python
#I need to normalize the data so that all the statistics can be analyzed on the same scale.
least_points = pd.Series(281*[min(Player_df['PTS'])])
least_steals = pd.Series(281*[min(Player_df['STL'])])
least_assists = pd.Series(281*[min(Player_df['AST'])])
least_rebounds = pd.Series(281*[min(Player_df['REB'])])
least_blocks = pd.Series(281*[min(Player_df['BLK'])])


#Equation to normalize each column from the original dataframe: (data-min(data)/(max(data)-min(data)))
def normalize_col(column, minimum):
    return (column-minimum)/(max(column)-min(column))



normalized_pts_column = normalize_col(Player_df['PTS'], least_points)
normalized_ast_column = normalize_col(Player_df['AST'], least_assists)
normalized_reb_column = normalize_col(Player_df['REB'], least_rebounds)
normalized_stl_column = normalize_col(Player_df['STL'], least_steals)
normalized_blk_column = normalize_col(Player_df['BLK'], least_blocks)


#Now finally let's make the normalized dataframe.
Normalized_df = pd.DataFrame()

def create_table(position, name, column):
    Normalized_df.insert(position, column=name, value=column)
    
    

create_table(0, 'Name', Player_df['name'])
create_table(1, 'Norm_PTS', normalized_pts_column)
create_table(2, 'Norm_AST', normalized_ast_column)
create_table(3, 'Norm_REB', normalized_reb_column)
create_table(4, 'Norm_STL', normalized_stl_column)
create_table(5, 'Norm_BLK', normalized_blk_column)


print(Normalized_df)
```

```python
#Create a list of player centroids and dataframe for visualization purposes.

rows = [['Well Rounded/Elite Player', 26, 6, 6.2, 1.4, 1.1], ['Scorer', 24, 2.8, 3, 0.7, 0.3], 
        ['Floor General', 16.2, 7.9, 5, 1.5, 1], ['Rebounder', 15, 1.5, 12, 0.5, 1.1], 
        ['Lockdown Defender',14.1, 3, 8, 2.3, 1.3], ['Average Player', 14.5, 6, 4.2, 0.8, 0.1]]
cols = ['type', 'PTS', 'AST', 'REB', 'STL', 'BLK']


player_type_df = pd.DataFrame(rows, columns=cols)
print(player_type_df)
```

```python
min_points = pd.Series(6*[min(Player_df['PTS'])])
min_steals = pd.Series(6*[min(Player_df['STL'])])
min_assists = pd.Series(6*[min(Player_df['AST'])])
min_rebounds = pd.Series(6*[min(Player_df['REB'])])
min_blocks = pd.Series(6*[min(Player_df['BLK'])])

max_points = pd.Series(6*[max(Player_df['PTS'])])
max_assists = pd.Series(6*[max(Player_df['AST'])])
max_rebounds = pd.Series(6*[max(Player_df['REB'])])
max_blocks = pd.Series(6*[max(Player_df['BLK'])])
max_steals = pd.Series(6*[max(Player_df['STL'])])

#Modified and reused the normalize column function from earlier
def normalize_centroids(column, minimum, maximum):
    return (column-minimum)/(maximum-minimum)

normalized_centroid_pts = normalize_centroids(player_type_df['PTS'], min_points, max_points)
normalized_centroid_ast = normalize_centroids(player_type_df['AST'], min_assists, max_assists)
normalized_centroid_reb = normalize_centroids(player_type_df['REB'], min_rebounds, max_rebounds)
normalized_centroid_stl = normalize_centroids(player_type_df['STL'], min_steals, max_steals)
normalized_centroid_blk = normalize_centroids(player_type_df['BLK'], min_blocks, max_blocks)

normalized_centroid_df = pd.DataFrame()

#Modified and reused the create table function from earlier.
def create_centroid_table(position, name, column):
    normalized_centroid_df.insert(position, column=name, value=column)

create_centroid_table(0, 'Name', player_type_df['type'])
create_centroid_table(1, 'PTS', normalized_centroid_pts)
create_centroid_table(2, 'AST', normalized_centroid_ast)
create_centroid_table(3, 'REB', normalized_centroid_reb)
create_centroid_table(4, 'STL', normalized_centroid_stl)
create_centroid_table(5, 'BLK', normalized_centroid_blk)

print(normalized_centroid_df)
```

```python
"""Initialize the KMeans classification with 6 clusters for the 6 player labels.
Use the normalized_centroid data as the positions for those clusters. Finally,
run the kmeans algorithm and take a look at the results."""
kmeans_model = KMeans(n_clusters=6, init=normalized_centroid_df.iloc[:,1:],random_state=1).fit(Normalized_df.iloc[:, 1:])

labels = kmeans_model.labels_
#Out of curiosity, I wanted to see what the final centroids would shape up to look like.
centroids = kmeans_model.cluster_centers_


results = pd.crosstab(labels,Normalized_df["Name"])
print(results)
print(centroids)
```
